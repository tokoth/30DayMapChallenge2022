```python
import geopandas as gpd
import folium
from folium.plugins import MarkerCluster
# File path
points_fp = r"pubs.shp" 
points_fp2= r"food.shp"
# Read the data
drink = gpd.read_file(points_fp)
food = gpd.read_file(points_fp2)
#Check input data
drink.head()
```

```python
food.head()
food["x"] = food["geometry"].x
food["y"] = food["geometry"].y
foods = list(zip(food["y"], food["x"]))
foods
```

```python
# Create a Map instance
m = folium.Map(location=[-1.2987,36.8185], tiles = 'OpenStreetMap', zoom_start=13, control_scale=True)
# Get x and y coordinates for each point in drink
drink["x"] = drink["geometry"].x
drink["y"] = drink["geometry"].y

# Get x and y coordinates for each point in food
#food["x"] = food["geometry"].x
#food["y"] = food["geometry"].y

#Add Title to Interactive map
loc = 'Get a Food or Drink in Nairobi'
title_html = '''{}'''.format(loc) 
m.get_root().html.add_child(folium.Element(title_html))

# Create a list of coordinate pairs
drinks = list(zip(drink["y"], drink["x"]))
#foods = list(zip(food["y"], food["x"]))

#Create feature groups
f1=folium.FeatureGroup("Coffee, Beer or Drink")
f2=folium.FeatureGroup("Restaraunt, Hotel or Fast Food")

# Create a folium marker cluster
marker_cluster = MarkerCluster(drinks).add_to(f1)
marker_cluster2 = MarkerCluster(foods).add_to(f2)

# Add marker cluster to map
f1.add_to(m)
f2.add_to(m)
folium.LayerControl().add_to(m)
outfp = r"day15_interactive_map.html"

m.save(outfp)

m
```
